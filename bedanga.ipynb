```python
import cv2
import os
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread("mypic.jpg")  # Replace with the correct image path

# Get the secret message and password
msg = input("Enter secret message: ")
password = input("Enter a passcode: ")

# Ensure that the message only contains printable characters (ASCII values 32-126)
msg = ''.join([c for c in msg if 32 <= ord(c) <= 126])

# Create dictionaries for character encoding and decoding
d = {}
c = {}

for i in range(32, 127):  # Only use printable ASCII characters
    d[chr(i)] = i
    c[i] = chr(i)

m = 0
n = 0
z = 0

# Encrypt the image with the secret message
for i in range(len(msg)):
    img[n, m, z] = d[msg[i]]
    n = n + 1
    m = m + 1
    z = (z + 1) % 3

# Add a marker for the end of the message, like a special character (e.g., ASCII value 0)
img[n, m, z] = 0  # End marker (ensure we don't overwrite the last message char)

# Save the encrypted image
cv2.imwrite("encryptedImage.jpg", img)

# Display the encrypted image in Colab
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert from BGR to RGB
plt.axis('off')  # Hide axes
plt.show()

# Save the password for future decryption
with open('password.txt', 'w') as f:
    f.write(password)

print("Encryption completed!")
```
